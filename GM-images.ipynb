```python
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException, TimeoutException
import pandas as pd
import time
from bs4 import BeautifulSoup as BS
import re
```

```python
driver = webdriver.Chrome()
driver.maximize_window()

df1 = pd.read_excel(r"C:\Users\panka\OneDrive\Desktop\OverAll India\Bots\Cleaned 19K.xlsx")
```

```python
driver = webdriver.Chrome()
driver.maximize_window()


for i in df1["Google Map Link"][18199:18600]:
    driver.get(i)
    try:
      
        img_button = WebDriverWait(driver, 5).until(EC.element_to_be_clickable((By.XPATH, '//button[@jsaction="pane.heroHeaderImage.click"]')))
        img_button.click()

        time.sleep(2)  

        pattern = r'url\("([^"]+)"\)'

        az = driver.find_elements(by="xpath", value ='//div[@class="m6QErb "]/div/div/a/div')

        for j in az[0:6]:
            ss = j.get_attribute("style")
            matches = re.search(pattern, ss)

            if matches:
                url = matches.group(1)
             
                new_url = re.sub(r'=w\d+-h\d+-k-no', '=w1200-h1000-k-no', url)

                GM_Link.append(i)
                IMG_Links.append(new_url)
    except TimeoutException as e:

        new_url = i 
        GM_Link.append(i)
        IMG_Links.append(new_url)

    except NoSuchElementException as e:

        new_url = i 
        GM_Link.append(i)
        IMG_Links.append(new_url)

driver.quit()
```

```python
dfz = pd.DataFrame()

dfz["Google Map Link"] = GM_Link
dfz["Image Links"] = IMG_Links
```
